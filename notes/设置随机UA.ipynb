# 设置随机UA
## 使用fake-useragent库生成随机UA
```shell
pip install fake-useragent
```

````python
import json

from fake_useragent import UserAgent

ua = UserAgent()

# 随机UA
print("随机UA:",ua.random)
# 指定浏览器类型的UA
print("chrome UA:",ua.chrome)
print("firefox UA:",ua.firefox)
````

## requests中使用随机UA


````python
import json
import requests

url = "https://httpbin.org/get"

headers = {
    "User-Agent": ua.random
}

res = requests.get(url, headers=headers).json()
print(json.dumps(res))
````
